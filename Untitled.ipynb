```python
import pandas as pd
```

```python
data3 = open("./carclassification1.txt",'r')
```

```python
data3
```

```python
columnas=data3.readline().strip().split(",")
```

```python
num_columnas= len(columnas)
```

```python
num_columnas
```

```python
contador=0
main_dic= {}
for columnas1 in columnas:
    main_dic[columnas1]=[]
```

```python
main_dic
```

```python
for line in data3:
    values= line.strip().split(",")
    for i in range(len(columnas)):
        main_dic[columnas[i]].append(values[i])
        contador += 1

print("El data set tiene %d filas y %d columnas"%(contador, num_columnas))
```

```python
main_dic
```

```python
df1=pd.DataFrame(main_dic)
```
